```python
def get_valid(data,year,month,player_name):
    valid = pd.DataFrame()
    data_ = data[(data['year'] == year) & (data['month'] == month)]
    data_ = data_.rename(columns = {'pts_x':'pts'})
    for i in data_.columns:
        if 'mp' in i and data_[i].dtype != 'float64':
            if len(data_[i].unique()) != 1:
                data_[i] = data_[i].apply(lambda x: float(x.replace(':','.')))    
    for i in data_.columns:
        if 'player' in i:
            idx = data_.columns.get_loc(i)
            if len(data_[data_[i].str.contains(player_name)].iloc[:,idx: (idx + 21)]) != 0:
                gp = len(data_[data_[i].str.contains(player_name)].iloc[:,idx: (idx + 21)])
                value = data_[data_[i].str.contains(player_name)].iloc[:,idx: (idx + 21)]
        
                value = pd.DataFrame(value.iloc[:,1:].sum()).T

                col = []
                for v in value.columns:
#                     if 'year' not in v or ''
                    k = v.partition('.')[0]
#                     value = value.rename(columns = {v:k})
                    col.append(k)
                value.columns = col        
                value['gp'] = gp
                value['player'] = player_name
                valid = pd.concat([valid,value])       
    return valid
```

```python
def first_few_month(data,month_num_):
    player_month = pd.DataFrame()
    for yr in data['year'].unique():
        n = []
        for i in data[data['year'] == yr].columns:
            if 'player' in i :
                for j in data[i].unique():
                    if j not in n:
                        n.append(j)
        for j in n :
            d = pd.DataFrame()
            for m in uni_month[:month_num_]:
                valid = get_valid(data,yr,m,j).reset_index(drop = True)
                if len(valid) == 0:
                    col_ = player_month.columns[:-1]
                    valid = pd.DataFrame(columns = col_)
                    valid['player'] = pd.Series(j)
                d = pd.concat((d,valid)).fillna(0)

            if 'gp' in d.columns:
                if d['gp'].sum() == 0:
                    d = pd.DataFrame(d.iloc[0]).T
                else:
                    d = pd.DataFrame(d.iloc[:,:-1].sum()/d['gp'].sum()).T
            d['player'] = j
            d['year'] = yr
            player_month = pd.concat((player_month,d))
    return player_month
# player_month = first_few_month(regular,2)
# player_month.to_csv('player_first_2_month.csv',index = False)
```
